```python
# =============================================================================
# Documento 1: Lista Simplemente Enlazada (LSE) con Sentinela
# =============================================================================

class SNode:
    """
    Representa un nodo para una Lista Simplemente Enlazada.
    Almacena un valor y una referencia al siguiente nodo.
    """
    def __init__(self, value=None, next_node=None):
        self.value = value
        self.next = next_node

    def __repr__(self):
        return f"SNode(value={self.value})"


class SinglyLinkedList:
    """
    Implementación de una Lista Simplemente Enlazada (LSE) que utiliza
    un nodo cabecera (sentinela) para simplificar las operaciones.
    """
    def __init__(self):
        """
        Constructor. Inicializa la lista con un nodo sentinela.
        La lista está inicialmente vacía.
        """
        # El nodo sentinela no contiene un valor útil.
        # Su 'next' apunta al primer elemento real de la lista.
        self.head = SNode(value="SENTINEL")
        self.size = 0

    def __len__(self):
        """Devuelve el número de elementos en la lista."""
        return self.size

    def __repr__(self):
        """
        Crea una representación en string de la lista para imprimirla.
        Ej: "7 -> 12 -> 3"
        """
        nodes = []
        current = self.head.next
        while current:
            nodes.append(str(current.value))
            current = current.next

        if not nodes:
            return "Lista vacía"
        return " -> ".join(nodes)

    def insert_head(self, value):
        """
        Inserta un nuevo elemento al principio de la lista.
        La operación es de tiempo constante, O(1).
        """
        # 1. Crear el nuevo nodo.
        new_node = SNode(value)

        # 2. El 'next' del nuevo nodo apunta al que era el primer nodo.
        new_node.next = self.head.next

        # 3. El 'next' del sentinela ahora apunta al nuevo nodo.
        self.head.next = new_node

        # 4. Incrementar el tamaño.
        self.size += 1
        print(f"-> Se insertó '{value}'. Estado de la lista: {self}")

    def search(self, key):
        """
        Busca un valor (key) en la lista.
        Devuelve el NODO que contiene el valor si se encuentra, de lo contrario None.
        La operación es de tiempo lineal en el peor caso, O(n).
        """
        print(f"\n? Buscando el valor '{key}'...")
        current = self.head.next  # Empezar desde el primer nodo real

        while current:
            if current.value == key:
                print(f"   ¡Encontrado! El nodo es {current}.")
                return current
            current = current.next

        print(f"   El valor '{key}' no se encontró en la lista.")
        return None

    def delete(self, key):
        """
        Elimina el primer nodo que contenga el valor (key).
        La operación es de tiempo lineal en el peor caso, O(n), debido a la búsqueda.
        """
        print(f"\n- Intentando eliminar el valor '{key}'...")
        prev = self.head
        current = self.head.next

        while current:
            if current.value == key:
                # Se encontró el nodo a eliminar (current).
                # El 'next' del nodo anterior (prev) ahora apunta al 'next' del nodo actual,
                # "saltándose" el nodo a eliminar.
                prev.next = current.next
                self.size -= 1
                print(f"   ¡Eliminado! Estado de la lista: {self}")
                return True # Indicar que la eliminación fue exitosa

            # Avanzar ambos punteros
            prev = current
            current = current.next

        print(f"   No se pudo eliminar: el valor '{key}' no existe.")
        return False


# =============================================================================
# SECCIÓN DE EJEMPLOS DE USO (LSE)
# =============================================================================
if __name__ == "__main__":
    print("--- INICIO: Pruebas con Lista Simplemente Enlazada (LSE) ---")

    # 1. Crear una nueva lista
    lse = SinglyLinkedList()
    print(f"Lista recién creada: {lse} (Tamaño: {len(lse)})")

    # 2. Insertar elementos
    print("\n--- Insertando elementos ---")
    lse.insert_head(10)
    lse.insert_head(20)
    lse.insert_head("A")
    lse.insert_head(30)

    # 3. Buscar elementos
    print("\n--- Buscando elementos ---")
    lse.search("A")      # Buscar un elemento existente
    lse.search(99)     # Buscar un elemento que no existe
    lse.search(10)     # Buscar el último elemento

    # 4. Eliminar elementos
    print("\n--- Eliminando elementos ---")
    lse.delete("A")      # Eliminar un elemento del medio
    lse.delete(99)     # Intentar eliminar un elemento que no existe
    lse.delete(30)     # Eliminar el primer elemento
    lse.delete(10)     # Eliminar el último elemento
    lse.delete(20)     # Vaciar la lista

    print(f"\nEstado final de la lista: {lse} (Tamaño: {len(lse)})")
    print("--- FIN: Pruebas LSE ---")
```

```python
# =============================================================================
# Documento 2: Lista Doblemente Enlazada (LDE) con Sentinela
# =============================================================================

class DNode:
    """
    Representa un nodo para una Lista Doblemente Enlazada.
    Almacena un valor y referencias al nodo previo y al siguiente.
    """
    def __init__(self, value=None, prev_node=None, next_node=None):
        self.value = value
        self.prev = prev_node
        self.next = next_node

    def __repr__(self):
        return f"DNode(value={self.value})"


class DoublyLinkedList:
    """
    Implementación de una Lista Doblemente Enlazada (LDE) que utiliza
    un nodo cabecera (sentinela).
    """
    def __init__(self):
        """
        Constructor. Inicializa la lista con un nodo sentinela.
        El 'prev' del sentinela siempre es None en una LDE no circular.
        """
        self.head = DNode(value="SENTINEL")
        self.size = 0

    def __len__(self):
        """Devuelve el número de elementos en la lista."""
        return self.size

    def __repr__(self):
        """
        Crea una representación en string de la lista para imprimirla.
        Ej: "30 <-> A <-> 10"
        """
        nodes = []
        current = self.head.next
        while current:
            nodes.append(str(current.value))
            current = current.next

        if not nodes:
            return "Lista vacía"
        return " <-> ".join(nodes)

    def insert_head(self, value):
        """
        Inserta un nuevo elemento al principio de la lista.
        La operación es de tiempo constante, O(1).
        """
        # 1. Crear el nuevo nodo y enlazarlo con el sentinela y el antiguo primer nodo.
        old_first = self.head.next
        new_node = DNode(value=value, prev_node=self.head, next_node=old_first)

        # 2. Actualizar el 'next' del sentinela para que apunte al nuevo nodo.
        self.head.next = new_node

        # 3. Si la lista no estaba vacía, actualizar el 'prev' del antiguo primer nodo.
        if old_first:
            old_first.prev = new_node

        self.size += 1
        print(f"-> Se insertó '{value}'. Estado de la lista: {self}")

    def search(self, key):
        """
        Busca un valor (key) en la lista.
        Devuelve el NODO que contiene el valor si se encuentra, de lo contrario None.
        Operación de tiempo lineal, O(n).
        """
        print(f"\n? Buscando el valor '{key}'...")
        current = self.head.next

        while current:
            if current.value == key:
                print(f"   ¡Encontrado! El nodo es {current}.")
                return current
            current = current.next

        print(f"   El valor '{key}' no se encontró en la lista.")
        return None

    def delete(self, key):
        """
        Elimina el primer nodo que contenga el valor (key).
        La operación es O(n) por la búsqueda, pero la eliminación en sí es O(1)
        una vez que se encuentra el nodo.
        """
        print(f"\n- Intentando eliminar el valor '{key}'...")
        node_to_delete = self.search(key) # Reutilizamos la búsqueda

        if node_to_delete:
            prev_node = node_to_delete.prev
            next_node = node_to_delete.next

            # El 'next' del nodo previo ahora apunta al 'next' del nodo a eliminar.
            prev_node.next = next_node

            # Si no es el último nodo, el 'prev' del nodo siguiente apunta al 'prev' del nodo a eliminar.
            if next_node:
                next_node.prev = prev_node

            self.size -= 1
            print(f"   ¡Eliminado! Estado de la lista: {self}")
            return True
        else:
            # El mensaje de "no encontrado" ya lo imprime la función search.
            print(f"   No se pudo eliminar.")
            return False


# =============================================================================
# SECCIÓN DE EJEMPLOS DE USO (LDE)
# =============================================================================
if __name__ == "__main__":
    print("--- INICIO: Pruebas con Lista Doblemente Enlazada (LDE) ---")

    # 1. Crear una nueva lista
    lde = DoublyLinkedList()
    print(f"Lista recién creada: {lde} (Tamaño: {len(lde)})")

    # 2. Insertar elementos
    print("\n--- Insertando elementos ---")
    lde.insert_head(10)
    lde.insert_head(20)
    lde.insert_head("A")
    lde.insert_head(30)

    # 3. Buscar elementos
    print("\n--- Buscando elementos ---")
    lde.search("A")      # Buscar un elemento existente
    lde.search(99)     # Buscar un elemento que no existe

    # 4. Eliminar elementos
    print("\n--- Eliminando elementos ---")
    lde.delete("A")      # Eliminar un elemento del medio
    lde.delete(99)     # Intentar eliminar un elemento que no existe
    lde.delete(30)     # Eliminar el primer elemento
    lde.delete(10)     # Eliminar el último elemento
    lde.delete(20)     # Vaciar la lista

    print(f"\nEstado final de la lista: {lde} (Tamaño: {len(lde)})")
    print("--- FIN: Pruebas LDE ---")
```

```python
# =============================================================================
# Documento 3: Lista Doblemente Enlazada Circular (LDEC) con Sentinela
# =============================================================================

class DNode:
    """
    Representa un nodo para una Lista Doblemente Enlazada.
    Se reutiliza para la versión circular.
    """
    def __init__(self, value=None, prev_node=None, next_node=None):
        self.value = value
        self.prev = prev_node
        self.next = next_node

    def __repr__(self):
        return f"DNode(value={self.value})"


class CircularDoublyLinkedList:
    """
    Implementación de una Lista Doblemente Enlazada Circular (LDEC)
    que utiliza un nodo cabecera (sentinela).
    """
    def __init__(self):
        """
        Constructor. El sentinela se apunta a sí mismo en una
        lista circular vacía.
        """
        self.head = DNode(value="SENTINEL")
        self.head.next = self.head
        self.head.prev = self.head
        self.size = 0

    def __len__(self):
        """Devuelve el número de elementos en la lista."""
        return self.size

    def __repr__(self):
        """
        Crea una representación en string de la lista.
        El recorrido se detiene cuando vuelve al nodo sentinela.
        """
        nodes = []
        # Empezamos en el primer nodo real
        current = self.head.next
        # Iteramos hasta que volvamos al sentinela
        while current != self.head:
            nodes.append(str(current.value))
            current = current.next

        if not nodes:
            return "Lista vacía"
        return " <-> ".join(nodes)

    def insert_head(self, value):
        """Inserta un nuevo elemento al principio de la lista (después del sentinela). O(1)."""
        old_first = self.head.next
        # 1. Crear el nuevo nodo y enlazarlo entre el sentinela y el antiguo primer nodo.
        new_node = DNode(value=value, prev_node=self.head, next_node=old_first)

        # 2. El 'prev' del antiguo primer nodo apunta ahora al nuevo nodo.
        old_first.prev = new_node

        # 3. El 'next' del sentinela apunta al nuevo nodo.
        self.head.next = new_node

        self.size += 1
        print(f"-> Se insertó '{value}' al inicio. Estado de la lista: {self}")

    def insert_tail(self, value):
        """Inserta un nuevo elemento al final de la lista (antes del sentinela). O(1)."""
        old_last = self.head.prev
        # 1. Crear el nuevo nodo y enlazarlo entre el antiguo último y el sentinela.
        new_node = DNode(value=value, prev_node=old_last, next_node=self.head)

        # 2. El 'next' del antiguo último nodo ahora apunta al nuevo nodo.
        old_last.next = new_node

        # 3. El 'prev' del sentinela apunta al nuevo nodo.
        self.head.prev = new_node

        self.size += 1
        print(f"-> Se insertó '{value}' al final.  Estado de la lista: {self}")


    def search(self, key):
        """
        Busca un valor (key) en la lista.
        Devuelve el NODO si se encuentra, de lo contrario None. O(n).
        """
        print(f"\n? Buscando el valor '{key}'...")
        current = self.head.next

        while current != self.head:
            if current.value == key:
                print(f"   ¡Encontrado! El nodo es {current}.")
                return current
            current = current.next

        print(f"   El valor '{key}' no se encontró en la lista.")
        return None

    def delete(self, key):
        """
        Elimina el primer nodo que contenga el valor (key). O(n).
        """
        print(f"\n- Intentando eliminar el valor '{key}'...")
        node_to_delete = self.search(key)

        if node_to_delete:
            prev_node = node_to_delete.prev
            next_node = node_to_delete.next

            # Desenlazar el nodo de la lista
            prev_node.next = next_node
            next_node.prev = prev_node

            self.size -= 1
            print(f"   ¡Eliminado! Estado de la lista: {self}")
            return True
        else:
            print(f"   No se pudo eliminar.")
            return False


# =============================================================================
# SECCIÓN DE EJEMPLOS DE USO (LDEC)
# =============================================================================
if __name__ == "__main__":
    print("--- INICIO: Pruebas con Lista Doblemente Enlazada Circular (LDEC) ---")

    # 1. Crear una nueva lista
    ldec = CircularDoublyLinkedList()
    print(f"Lista recién creada: {ldec} (Tamaño: {len(ldec)})")

    # 2. Insertar elementos
    print("\n--- Insertando elementos ---")
    ldec.insert_head(10)
    ldec.insert_head(20)
    ldec.insert_tail("Z") # ¡Inserción al final en O(1)!
    ldec.insert_head(30)

    # 3. Buscar elementos
    print("\n--- Buscando elementos ---")
    ldec.search("Z")      # Buscar un elemento existente
    ldec.search(99)     # Buscar un elemento que no existe

    # 4. Eliminar elementos
    print("\n--- Eliminando elementos ---")
    ldec.delete(20)     # Eliminar un elemento del medio
    ldec.delete(99)     # Intentar eliminar un elemento que no existe
    ldec.delete(30)     # Eliminar el primer elemento
    ldec.delete("Z")    # Eliminar el último elemento
    ldec.delete(10)     # Vaciar la lista

    print(f"\nEstado final de la lista: {ldec} (Tamaño: {len(ldec)})")
    print("--- FIN: Pruebas LDEC ---")
```

```python
# =============================================================================
# Bloque 1: Implementación de las Estructuras de Datos
# Se han unido las tres clases y se han eliminado los mensajes de print()
# en las operaciones para no afectar las mediciones de tiempo,
# como se recomienda en el taller.
# =============================================================================

import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Clases para Lista Simplemente Enlazada (LSE) ---
class SNode:
    def __init__(self, value=None, next_node=None):
        self.value = value
        self.next = next_node

class SinglyLinkedList:
    def __init__(self):
        self.head = SNode(value="SENTINEL")
        self.size = 0

    def __len__(self):
        return self.size

    def __repr__(self):
        nodes = []
        current = self.head.next
        while current:
            nodes.append(str(current.value))
            current = current.next
        return " -> ".join(nodes) if nodes else "Lista vacía"

    def insert_head(self, value):
        new_node = SNode(value, self.head.next)
        self.head.next = new_node
        self.size += 1

    def search(self, key):
        current = self.head.next
        while current:
            if current.value == key:
                return current
            current = current.next
        return None

    def delete(self, key):
        prev = self.head
        current = self.head.next
        while current:
            if current.value == key:
                prev.next = current.next
                self.size -= 1
                return True
            prev = current
            current = current.next
        return False

# --- Clases para Listas Doblemente Enlazadas (LDE y LDEC) ---
class DNode:
    def __init__(self, value=None, prev_node=None, next_node=None):
        self.value = value
        self.prev = prev_node
        self.next = next_node

class DoublyLinkedList:
    def __init__(self):
        self.head = DNode(value="SENTINEL")
        self.size = 0

    def __len__(self):
        return self.size

    def __repr__(self):
        nodes = []
        current = self.head.next
        while current:
            nodes.append(str(current.value))
            current = current.next
        return " <-> ".join(nodes) if nodes else "Lista vacía"

    def insert_head(self, value):
        old_first = self.head.next
        new_node = DNode(value=value, prev_node=self.head, next_node=old_first)
        self.head.next = new_node
        if old_first:
            old_first.prev = new_node
        self.size += 1

    def search(self, key):
        current = self.head.next
        while current:
            if current.value == key:
                return current
            current = current.next
        return None

    def delete(self, key):
        node_to_delete = self.search(key)
        if node_to_delete:
            prev_node = node_to_delete.prev
            next_node = node_to_delete.next
            prev_node.next = next_node
            if next_node:
                next_node.prev = prev_node
            self.size -= 1
            return True
        return False

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = DNode(value="SENTINEL")
        self.head.next = self.head
        self.head.prev = self.head
        self.size = 0

    def __len__(self):
        return self.size

    def __repr__(self):
        nodes = []
        current = self.head.next
        while current != self.head:
            nodes.append(str(current.value))
            current = current.next
        return " <-> ".join(nodes) if nodes else "Lista vacía"

    def insert_head(self, value):
        old_first = self.head.next
        new_node = DNode(value=value, prev_node=self.head, next_node=old_first)
        old_first.prev = new_node
        self.head.next = new_node
        self.size += 1

    def insert_tail(self, value):
        old_last = self.head.prev
        new_node = DNode(value=value, prev_node=old_last, next_node=self.head)
        old_last.next = new_node
        self.head.prev = new_node
        self.size += 1

    def search(self, key):
        current = self.head.next
        while current != self.head:
            if current.value == key:
                return current
            current = current.next
        return None

    def delete(self, key):
        node_to_delete = self.search(key)
        if node_to_delete and node_to_delete != self.head:
            prev_node = node_to_delete.prev
            next_node = node_to_delete.next
            prev_node.next = next_node
            next_node.prev = prev_node
            self.size -= 1
            return True
        return False


# =============================================================================
# Bloque 2: Función de Medición (sugerida en el taller)
# =============================================================================
def bench(fn, *args, repeats=7, warmup=2, **kwargs):
    for _ in range(warmup):
        fn(*args, **kwargs)
    times = []
    for _ in range(repeats):
        t0 = time.perf_counter()
        fn(*args, **kwargs)
        t1 = time.perf_counter()
        times.append((t1 - t0) * 1000)  # Convertir a milisegundos
    times.sort()
    return times[len(times) // 2]


# =============================================================================
# Bloque 3: Ejecución de los Experimentos
# =============================================================================
results = []
list_sizes = [1000, 5000, 10000, 20000, 40000]
list_implementations = {
    "LSE": SinglyLinkedList,
    "LDE": DoublyLinkedList,
    "LDEC": CircularDoublyLinkedList
}

for name, ListClass in list_implementations.items():
    for n in list_sizes:
        print(f"Probando {name} con n={n}...")

        # --- PREPARACIÓN DE DATOS ---
        lista = ListClass()
        # Llenar la lista con n elementos (de n-1 a 0)
        for i in range(n):
            lista.insert_head(i)

        key_best = n - 1  # Primer elemento, búsqueda/eliminación más rápida
        key_worst = 0     # Último elemento, búsqueda/eliminación más lenta

        # --- MEDICIÓN DE OPERACIONES ---

        # Insert Head (siempre Θ(1))
        time_insert = bench(lista.insert_head, value="new")
        results.append({"Tipo": name, "Operación": "insert_head", "Caso": "N/A", "Tamaño (n)": n, "Tiempo (ms)": time_insert})

        # Insert Tail (solo para LDEC)
        if name == "LDEC":
            time_insert_tail = bench(lista.insert_tail, value="new_tail")
            results.append({"Tipo": name, "Operación": "insert_tail", "Caso": "N/A", "Tamaño (n)": n, "Tiempo (ms)": time_insert_tail})

        # Search
        time_search_best = bench(lista.search, key=key_best)
        results.append({"Tipo": name, "Operación": "search", "Caso": "Mejor", "Tamaño (n)": n, "Tiempo (ms)": time_search_best})
        time_search_worst = bench(lista.search, key=key_worst)
        results.append({"Tipo": name, "Operación": "search", "Caso": "Peor", "Tamaño (n)": n, "Tiempo (ms)": time_search_worst})

        # Delete
        time_delete_best = bench(lista.delete, key=key_best)
        results.append({"Tipo": name, "Operación": "delete", "Caso": "Mejor", "Tamaño (n)": n, "Tiempo (ms)": time_delete_best})
        lista.insert_head(key_best) # Reinsertar para la siguiente prueba

        time_delete_worst = bench(lista.delete, key=key_worst)
        results.append({"Tipo": name, "Operación": "delete", "Caso": "Peor", "Tamaño (n)": n, "Tiempo (ms)": time_delete_worst})

# Convertir resultados a un DataFrame de Pandas para fácil manipulación
df_results = pd.DataFrame(results)


# =============================================================================
# Bloque 4: Generación de Gráficas
# =============================================================================
sns.set_theme(style="whitegrid")
print("\nGenerando gráficas...")

# --- Gráfica 1: Comparación de Operaciones Lineales (Peor Caso) ---
plt.figure(figsize=(12, 7))
plot_data_linear = df_results[df_results["Caso"] == "Peor"]
sns.lineplot(data=plot_data_linear, x="Tamaño (n)", y="Tiempo (ms)", hue="Tipo", style="Operación", marker="o")
plt.title("Rendimiento de Operaciones Lineales (Peor Caso): T(n) vs n")
plt.xlabel("Tamaño de la Lista (n)")
plt.ylabel("Tiempo Mediano (ms)")
plt.legend(title="Estructura y Operación")
plt.show()

# --- Gráfica 2: Comparación de Operaciones Constantes (Θ(1)) ---
plt.figure(figsize=(12, 7))
plot_data_constant = df_results[
    (df_results["Caso"] == "Mejor") | (df_results["Caso"] == "N/A")
]
plot_data_constant["Etiqueta"] = plot_data_constant["Tipo"] + " - " + plot_data_constant["Operación"]
sns.lineplot(data=plot_data_constant, x="Tamaño (n)", y="Tiempo (ms)", hue="Etiqueta", marker="o")
plt.title("Rendimiento de Operaciones de Costo Constante (Θ(1))")
plt.xlabel("Tamaño de la Lista (n)")
plt.ylabel("Tiempo Mediano (ms)")
plt.legend(title="Operación", bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()
plt.show()
```
